Un primer intento  para ver si en el texto cifrado se encuentra uno de los trigramas más comunes **THE**, veamos si en el texto_cifrado, se encuentran la combinación de letras FIC, considerando a C = e Y F = t, según el diagrama de barras de arriba, e I = h.

```python
palabra = "FIC"

# Contamos cuántas veces aparece la secuencia "FIC" en el texto cifrado
conteo_palabra = texto_cifrado.count(palabra)

# Mostrar el resultado
conteo_palabra
```

La celda de codigo de arriba tenia la intención de buscar patrones de la palabra _FIC_, que en el idioma ingles podia ser **THE**, pero no se obtuvieron los resultados, Supongamos entonces que la letra I &ne; h 

```python
reemplazos = {
    'C': 'e',
    'F': 't',
}

# Reemplazar las letras según las suposiciones
texto_cifrado_cambio = texto_cifrado  # Iniciar con el texto original

for letra, reemplazo in reemplazos.items():
    texto_cifrado_cambio = texto_cifrado_cambio.replace(letra, reemplazo)  # Actualizar la variable

# Mostrar el texto resultante después de los reemplazos
print(texto_cifrado_cambio)  
```

Hemos reemplazado estas dos primeras letras, pero aún no es posible inferir nada del texto, a continuación vamos crear una función para encontrar n-gramas en el texto, que se repitan constantemente.

```python
def get_ngrams(text, n):
    """
    Genera n-gramas de longitud n a partir de un texto dado.
    """
    text = text.replace(" ", "").lower()  # Elimina espacios y pasa a minúsculas
    ngrams = [text[i:i+n] for i in range(len(text) - n + 1)]
    return ngrams

def analyze_ngrams(texto_cifrado, n, top=10):
    """
    Encuentra los n-gramas más frecuentes en el texto.
    """
    ngrams = get_ngrams(texto_cifrado, n)
    ngram_counts = Counter(ngrams)
    return ngram_counts.most_common(top)

top_trigrams = analyze_ngrams(texto_cifrado, 3, top=15)
top_quatrigrams = analyze_ngrams(texto_cifrado, 4, top=15)

top_digrams = analyze_ngrams(texto_cifrado, 2, top=15)

# Muestra los resultados
print("Top trigramas:")
for trigram, count in top_trigrams:
    print(f"{trigram}: {count}")

print("\nTop quatrigramas:")
for quatrigram, count in top_quatrigrams:
    print(f"{quatrigram}: {count}")

print("\nTop digramas:")
for digrams, count in top_digrams:
    print(f"{digrams}: {count}")
```

Continuando con nuestra suposición, y observando los resultados de los ngramas notamos que el trigrama fwc se repite 13 veces, este trigrama puede ser **THE**, asi , W = h.

El quatrigrama que contiene las letras f,w en el texto cifrado es fwvf, por la sintaxis del idioma ingles, Thot,thet,thit no eisten estás palabras, por lo tanto nuestra siguiente suposicion es que V = a

```python
quadgrams = {
    "C": "e",
    "F": "t",
    "W": "h",
    "V": "a"
    
}
def reemplazar_con_diccionario(diccionario, texto_cifrado):
    """
    Reemplaza las palabras clave del diccionario en el texto cifrado por sus valores.
    
    Args:
        diccionario (dict): Diccionario con pares llave-valor a usar para reemplazo.
        texto_cifrado (str): Texto cifrado donde se harán los reemplazos.
        
    Returns:
        str: Texto modificado con las palabras reemplazadas.
    """
    texto_modificado = texto_cifrado
    for llave, valor in diccionario.items():
        texto_modificado = texto_modificado.replace(llave, valor)
    return texto_modificado

texto_modificado = reemplazar_con_diccionario(quadgrams,texto_cifrado)
print(texto_modificado)
```
